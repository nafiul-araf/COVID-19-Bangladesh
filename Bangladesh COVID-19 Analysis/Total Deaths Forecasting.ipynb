```python
from statsmodels.tsa.stattools import adfuller
```

```python
def stationarity(data):
    result=adfuller(data)
    keys=['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    for key, value in zip(keys, result):
        print(key+' : '+str(value))
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")
```

```python
stationarity(df)
```

```python
df_shift=df-df.shift()
df_shift.dropna(inplace=True)
df_shift.head(8)
```

```python
df_shift['Total Deaths'].unique()
```

```python
stationarity(df_shift)
```

```python
df_log=np.log(df)
df_log.plot();
```

```python
df_log_shift=df_log-df_log.shift(2)
df_log_shift.dropna(inplace=True)
```

```python
df_log_shift['Total Deaths'].unique()
```

```python
df_log_shift=df_log_shift['Total Deaths'].replace([np.inf, -np.inf], np.nan)
df_log_shift
```

```python
df_log_shift=pd.DataFrame(df_log_shift)
df_log_shift
```

```python
df_log_shift.dropna(inplace=True)
df_log_shift['Total Deaths'].unique()
```

```python
plt.plot(df_log_shift, label='Original Log Shifted Data')
plt.plot(df_log_shift.rolling(7).mean(), color='red', label='Shifted Rolling Mean')
plt.plot(df_log_shift.rolling(7).std(), color='green', label='Shifted Rolling Standard Deviation')
plt.legend(loc='best')
plt.title('Shifted Rolling Mean and Average in Log Scale')
plt.tight_layout()
plt.show()
```

```python
stationarity(df_log_shift)
```

```python
add=seasonal_decompose(df_log_shift, model='additive', period=1)
add.plot().suptitle('\Additive Decompose in Log Shifted Data', fontsize=12);
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
ACF=acf(df_log_shift, nlags=15)
PACF=pacf(df_log_shift, nlags=15)

plt.subplot(121)
plt.plot(ACF, marker='+')
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=-1.96/np.sqrt(len(df_log_shift)), linestyle='--', color='gray')
plt.axhline(y=1.96/np.sqrt(len(df_log_shift)), linestyle='--', color='gray')
plt.title('Auto Correlation')
plt.xlabel('Lags')
plt.ylabel('Correlation')

plt.subplot(122)
plt.plot(PACF, marker='+')
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=-1.96/np.sqrt(len(df_log_shift)), linestyle='--', color='gray')
plt.axhline(y=1.96/np.sqrt(len(df_log_shift)), linestyle='--', color='gray')
plt.title('Partial Auto Correlation')
plt.xlabel('Lags')
plt.ylabel('Correlation')

plt.tight_layout()
plt.show()
```

```python
fig=plt.figure(figsize=(15, 10))
ax1=fig.add_subplot(211)
fig=plot_acf(df_log_shift, lags=15, ax=ax1)
ax2=fig.add_subplot(212)
fig=plot_pacf(df_log_shift, lags=15, ax=ax2)
plt.show()
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
model=ARIMA(df['Total Deaths'], order=(3, 1, 3))
model_fit=model.fit()
```

```python
print(model_fit.summary())
```

```python
residuals=pd.DataFrame(model_fit.resid)
residuals.plot()
plt.show()
```

```python
residuals.plot(kind='kde')
plt.show()
```

```python
df['forecast']=model_fit.predict(start=len(df)-22, end=len(df), dynamic=True)
df[['Total Deaths','forecast']].plot(figsize=(20,15))
plt.show()
```

```python
import statsmodels.api as sm
model=sm.tsa.statespace.SARIMAX(df['Total Deaths'], order=(3, 1, 3), seasonal_order=(1, 1, 1, 12))
results=model.fit()
```

```python
print(results.summary())
```

```python
residuals=pd.DataFrame(results.resid)
residuals.plot()
plt.show()
```

```python
residuals.plot(kind='kde')
plt.show()
```

```python
df['forecast']=results.predict(start=len(df)-22, end=len(df), dynamic=True)
df[['Total Deaths','forecast']].plot(figsize=(20,15))
plt.show()
```

```python
from pandas.tseries.offsets import DateOffset
future_dates=[df.index[-1]+ DateOffset(weeks=x)for x in range(0, 52)]
```

```python
future_datest_df=pd.DataFrame(index=future_dates[1:], columns=df.columns)
```

```python
future_df=pd.concat([df,future_datest_df])
```

```python
future_df['forecast']=results.predict(start=len(df)-1, end=len(df)+52, dynamic=True)  
future_df[['Total Deaths', 'forecast']].plot(figsize=(20, 15))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylabel('Deaths', fontsize=28)
plt.title('Total Deaths Forcasting upto September, 2022 in Bangladesh', fontsize=28)
plt.legend(loc='best', prop={'size': 18})
plt.show()
```
